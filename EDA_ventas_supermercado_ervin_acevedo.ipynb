## Transformaci贸n de los datos

### Validaci贸n de datos nulos y eliminaci贸n de datos nulos (en caso de nuevas entradas de registros)

### Uni贸n de registros coincidentes (inner join) para tener un solo archivo unificado

```python
df_con = (
    ventas
    .merge(inventarios, on='ID_Tienda')
    .merge(satisfaccion, on='ID_Tienda')
)
```

#### Se elimina la columna "Producto_y" debido a que el producto se puede referenciar por medio de la columna "Producto_x"

```python
df_con.drop(columns=['Producto_y'], inplace=True)
```

#### Pasamos todas las columnas restantes a lower_case para mejor manejo en el c贸digo

```python
df_con.columns = df_con.columns.str.lower()
```

## An谩lisis de inventario

### Stock total por tienda

```python
st_ti = df_con.groupby('id_tienda', observed = False) ['stock_disponible'].sum()
st_ti
```

### C谩lculo de inventario (rotaci贸n)

```python
df_con['rotacion_inventario'] = df_con['cantidad_vendida'] / df_con['stock_disponible']
df_con
```

### Stock total por tienda (rotaci贸n)

```python
st_ti_ro = df_con.groupby('id_tienda', observed = False) ['rotacion_inventario'].sum()
st_ti_ro
```

### Tiendas con inventario critico (< 10% de ventas respecto al stock)

```python
tl_stock = df_con.groupby('id_tienda')['stock_disponible'].sum()
tl_ventas = df_con.groupby('id_tienda')['cantidad_vendida'].sum()
por_ventas = (tl_stock / tl_ventas) * 100
td_criticas  = por_ventas[por_ventas < 10]
print("Tiendas con niveles cr铆ticos de inventario (< 10% de ventas respecto al stock):")
td_criticas
```
